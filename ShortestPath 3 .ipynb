```python
import networkx as nx
import matplotlib.pyplot as plt

# Define the nodes and their values
nodes = {
    'R': [7, 5, 0, 0],
    'S': [7, 0, 0, 2],
    'T': [0, 6, 0, 0],
    'U': [4, 0, 1, 0]
}

# Create a graph
G = nx.Graph()

# Add nodes to the graph
for node, values in nodes.items():
    G.add_node(node)

# Add edges based on the values
for node1, values1 in nodes.items():
    for node2, values2 in nodes.items():
        if node1 != node2:
            weight = sum([a*b for a, b in zip(values1, values2)])
            if weight > 0:
                G.add_edge(node1, node2, weight=weight)

# Plot the graph
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=700, node_color='skyblue', font_size=12, font_weight='bold')
nx.draw_networkx_edge_labels(G, pos, font_color='red', font_size=10)

# Get user input for source and target nodes
source_node = input("Enter source node: ").upper()
target_node = input("Enter target node: ").upper()

# Check if nodes are valid
if source_node not in nodes or target_node not in nodes:
    print("Invalid source or target node.")
else:
    # Calculate shortest path length
    shortest_path_length = nx.shortest_path_length(G, source=source_node, target=target_node)
    print(f"Length of shortest path from node {source_node} to node {target_node}: {shortest_path_length}")

    # Calculate density of the graph
    density = nx.density(G)
    print("Density of the graph:", density)

    # Save the graph as an image
    plt.title("Graph Visualization")
    plt.savefig("graph.png")
    plt.show()
```
